```python
# Carguemos modulos
import random as rnd
```

```python
# Dados  100 numeros aleatorios enteros en [666,999].
# Calcular cuantos son multiplos de 7

# NUmero de experimentos
num_exp = 500

# Almacenemos el numero de multiplos de 7
num_7k = 0

# Este for sirve para realizar un numero determinado de tareas
# Es comun la utilizacion de una estructura repetitiva for cuando se
# conoce el numero de veces que vamos a realizar una tarea determinada
for i in range(num_exp):
  # Genero un numero aleatorio con las condiciones del problema
  random1 = rnd.randint(666,999)
  # Validamos si el numero aleatorio recientemente generado es multiplo de 7
  if (random1 % 7 == 0):
    # actualizo el numero de multiplos de 7 generados de manera aleatoria
    num_7k = num_7k +1

# Mostremos los resultados
print("Numero de multiplos de 7 es : %i" %(num_7k))
```
