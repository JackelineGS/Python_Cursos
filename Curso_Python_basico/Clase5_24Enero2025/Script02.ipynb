```python
# Operacion de concatenacion
lista_4 = lista_3 + lista_2
lista_4
```

```python
# tipo de dato
type(lista_4)
```

```python
# Operacion de repeticion
lista_5 = 3*lista_4
lista_5
```

```python
# Funcion len
len(lista_6)
```
