```python
# Generemos una lista con 10**3 numeros aleatorios enteros en [0,20]

# Carguemos el modulo random
import random as rnd

lista_data = []
for i in range(10**3):
  lista_data.append(rnd.randint(0,20))

lista_data
```

```python
# Contemos las veces que aparece el numero 20
lista_data.count(20)
```

```python
# Contemos el numero de veces que aparece el -1
lista_data.count(-1)
```
