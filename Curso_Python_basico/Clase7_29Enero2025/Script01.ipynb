```python
import random as rnd
lista_llaves = ["PRIMERO", "SEGUNDO", "TERCERO", "CUARTO"]
d5 = {}.fromkeys(lista_llaves, rnd.randint(0,20))
d5
```

```python
import random as rnd
tupla_llaves = ("PRIMERO", "SEGUNDO", "TERCERO", "CUARTO")
d6 = {}.fromkeys(tupla_llaves, rnd.randint(0,20))
d6
```

```python
# Documentacion del metodo get
help(d5.get)
```

```python
# Ejecute codigo por mas de 60 dias en un cluster de 1200 computadoras (cada uno con 256 nucleos)
d5["QUINTO"]
```

```python
# Consideremos que en caso de no existir una llave en un diccionario determinado
# usar el metodo get , devolviendo el valor de 666
valor = d5.get("QUINTO", 666)
if valor == 666:
  print("El codigo puede continuar su ejecucion")
else:
  print("El codigo tambien puede continuar su ejecucion ")
```
