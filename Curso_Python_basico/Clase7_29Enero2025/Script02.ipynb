# Modulo Numpy

```python
# Carguemos el modulo numpy
import numpy as np
```

```python
# Documentacion del modulo
help(np)
```

```python
# # Numpy provee una serie de submodulos/subpaquetes

# random
#         Core Random Tools
#     linalg
#         Core Linear Algebra Tools
#     fft
#         Core FFT routines
#     polynomial
#         Polynomial tools
```

```python
# Elementos que componen al modulo numpu
dir(np)
```

```python
# Elementos que componen al modulo random
dir(np.random)
```

```python
# Documentacion del modulo random
help(np.random)
```

```python
# Documentacion de la funcion triangular del submodulo random del modulo numpy
help(np.random.triangular)
```

```python
# unidimensional
v1 = np.random.triangular(12,14,16,10)
v1
```

```python
# Bidimensional
v2 = np.random.triangular(12,14,16,(3,4))
v2
```

```python
# Tridimensional
v3 = np.random.triangular(12,14,16,(2,3,4 ))
v3
```

```python
# Antepenultima componente del objeto unidimensional v1
v1[-3]
```

```python
# Accedamos a elementos del objeto bidimensional v2
```

```python
# Tercera columna de v2
v2[:,2]
```

```python
# Tercera fila de v2
v2[2,:]
```

```python
# Mostremos la primera y ultima columna de v2
v2[:,(0,-1)]
```

```python
# Accedamos a elementos de un objeto tridimensional
v3[0,(1,2),1]
```
